# 금융 분야에서의 생성형 인공지능 (Generative AI)

## 1. 서문
- 최근 생성형 인공지능(GenAI)의 발전으로 금융 분야에서 AI 혁신에 대한 관심이 증가.
- 완전 자동화된 금융 시장에서 GenAI 사용은 아직 개발 단계에 있으며, 이를 통한 위험 또한 존재.

## 2. GenAI와 금융
- **GenAI와 전통적 AI의 차이점**: GenAI는 단순한 패턴 인식이 아닌 '원본' 콘텐츠를 생성할 수 있는 능력을 보유.
- **금융에서의 GenAI 사용 사례**:
  - 문서 요약, 번역 및 보고서 작성 지원.
  - 고객 서비스 분석 및 내부 운영 최적화.

## 3. 금융에서 GenAI가 제기하는 위험
- **공정성 문제**: 잘못된 데이터를 사용할 경우 편향성과 차별을 초래할 수 있음.
- **설명 가능성 부족**: AI 모델이 결과를 도출하는 방식 설명에 한계가 존재.
- **데이터 관련 위험**: 데이터 품질 저하, 프라이버시 침해, 데이터 집중화 가능성.
- **모델의 견고성**: 모델의 예측 능력 및 정확성 문제, 특히 시장 조작 가능성.
- **거버넌스 문제**: 책임과 투명성 부족, 외부 모델 의존에 따른 리스크.

## 4. 정책적 고려 사항
- **데이터 거버넌스 강화**: 편향성 및 차별 문제 해결을 위한 데이터 관리 강화 필요.
- **모델 투명성 향상**: AI 모델의 설명 가능성 및 투명성을 높이는 규제 필요.
- **모델 견고성 강화**: 사이버 보안 위험 완화 및 시장 조작 방지를 위한 모델 안전성 강화 필요.

# 1장: 생성형 인공지능(GenAI)과 금융

## 1.1 GenAI: 다른 유형의 AI와의 차이점
- **생성형 인공지능(GenAI)** 는 훈련 데이터에 기반해 새 콘텐츠를 생성할 수 있는 AI의 하위 범주입니다. 예를 들어, 대규모 언어 모델(LLM)은 언어/텍스트 생성에 특화된 GenAI의 한 유형입니다. 또한 GenAI는 시각적 출력(예: DALL-E), 오디오(예: Speechify), 이미지, 코드 생성에도 사용됩니다.
- GenAI는 대규모 비정형 데이터셋을 처리 및 학습하여 다양한 유형의 콘텐츠를 즉시 생성할 수 있으며, 주로 **생성적 적대 신경망(GAN)** 을 통해 이를 구현합니다.
- 기존 AI 모델은 주로 패턴 인식, 분류 및 예측을 위한 반면, GenAI는 사람과 거의 구분되지 않는 '원본' 콘텐츠를 생성할 수 있습니다.

### 1.1.1 주요 발전 요인
- GenAI 기술의 발전은 급격한 컴퓨팅 파워 향상과 데이터 처리 능력 증가에 기인합니다.
- **빅데이터** 와 소셜 미디어, 거래 데이터 등에서 방대한 데이터 포인트의 증가가 GenAI의 발전을 가속화하고 있습니다.

### 1.1.2 금융에서 AI의 느린 적용
- 금융 시장에서 완전 자동화된 AI 사용 사례는 대부분 실험 단계에 있으며, 현재 대부분의 AI는 **백오피스와 중간 관리**에서 **프로세스 자동화**를 통해 생산성을 높이는 데 사용됩니다.
- 금융 기관은 GenAI의 보안 및 데이터 주권 문제를 해결하기 위해 **폐쇄된 환경**에서 도구를 배포하는 방식으로 GenAI를 제한적으로 사용하고 있습니다.

### 1.1.3 AI의 직접적 및 간접적 사용 범위
- **간접적 사용**: AI가 금융 서비스 제공자가 고객에게 맞춤형 포트폴리오 할당 추천을 제공하는 데 도움을 줄 수 있습니다.  
- **직접적 사용**: 고객과의 직접 상호 작용을 허용하는 것은 위험하며, 고객의 재정적 위험을 증가시킬 수 있습니다.

## 1.2 금융에서 AI 및 GenAI의 활용 사례 및 이점
- AI 도구는 증권, 자산 관리, 은행 등 다양한 금융 시장에서 사용되고 있으며, 가치 사슬의 여러 부분에 걸쳐 적용됩니다.
- GenAI는 고객 지원, 보고서 작성, 위험 관리, 자산 관리, 사기 탐지, **AML/CFT**(자금세탁방지 및 테러 자금조달 방지)와 같은 분야에서 특히 유용합니다.
- **백오피스**에서 GenAI는 거래 후 처리, 손익 조정, 데이터 분석, 리스크 모델링, HR 관리 등에서 사용됩니다.
- **프론트오피스**에서는 고객 지원, 맞춤형 마케팅, 자산 배분, 고객 인증 등에 활용됩니다.

### 1.2.1 GenAI의 금융 분야 사용 사례 (Figure 1.6)
- GenAI는 백오피스 및 프론트오피스 모두에서 업무 최적화 및 자동화를 통해 운영 효율성을 크게 향상시킵니다.
- GenAI는 고객 세분화, 마케팅 전략 최적화, 금융 조언 제공, 코드 생성 등 다양한 분야에서 활용됩니다.

- **프론트오피스**: 고객 지원, 신상품 개발, 맞춤형 마케팅, 자산 배분, 고객 인증.

# 2장: 금융에서의 생성형 인공지능(GenAI) 리스크



## 2.1 공정성: 편향, 차별 및 불공정한 결과의 위험
- AI 모델은 데이터 품질이 낮거나 적절하지 않은 데이터를 사용하면 편향적이거나 차별적인 결과를 초래할 수 있음.
- 특히, GenAI는 온라인에서 이용할 수 있는 거의 모든 데이터를 학습할 수 있어 데이터에 내재된 편향을 강화할 위험이 큼.
- 예시로, 신용 대출 결정에서 AI 기반 모델이 차별적 대출 결정 위험을 야기할 수 있음.

## 2.2 설명 가능성 부족
- AI 모델이 어떻게 결과를 도출하는지 이해하는 것이 어려움. 이 문제는 GenAI에서 특히 심각.
- 금융에서 모델의 성능 저하나 운영 실패 시 설명 불가능성이 신뢰를 저해함.
- 예시로, 대출 거절 시 고객이 신용 개선 방법을 이해하거나 불공정 결정을 파악하기 어려움.

## 2.3 데이터 관련 위험: 데이터 품질, 프라이버시, 데이터 집중화
- AI 모델의 성능은 학습 데이터의 품질에 따라 결정되며, GenAI의 경우 사용자가 입력하는 데이터도 모델 성능에 영향을 미침.
- 데이터의 대표성 부족은 부정확한 모델 결과를 초래할 수 있으며, 데이터 프라이버시 문제도 존재.
- 온라인에서 데이터를 실시간으로 수집하는 플러그인 사용 시 개인정보 침해 가능성.

## 2.4 모델의 견고성 및 신뢰성, 시장 조작 위험
- GenAI는 잘못된 정보나 '환각'을 일으킬 수 있으며, 이는 금융 소비자에게 신뢰성 문제를 야기할 수 있음.
- 잘못된 투자 조언이나 부정확한 예측은 시장 조작이나 투자자의 피해를 초래할 수 있음.

## 2.5 거버넌스 관련 위험: 책임 및 투명성 부족
- AI 시스템의 관리 및 책임 소재가 불분명하면 문제 발생 시 해결이 어려움.
- GenAI 사용 시 제3자 의존도가 높아지고, 데이터 및 모델의 집중화로 시장의 시스템 리스크가 증가할 수 있음.

## 2.6 금융 안정성 관련 위험: 군집화, 변동성, 플래시 크래시
- GenAI 모델이 자율적으로 학습하고 조정하는 방식이 투자자 행동의 군집화를 초래하여 유동성 문제를 악화시킬 수 있음.
- 고빈도 거래에서의 시장 변동성 및 시스템 리스크가 더욱 심화될 수 있음.

## 2.7 경쟁 관련 위험
- GenAI 모델 개발에 필요한 컴퓨팅 자원과 데이터의 집중으로 인해 일부 대형 기업이 시장을 지배할 수 있음.
- 소규모 금융 기관이나 국가들은 이러한 기술에 접근하기 어려워져 경쟁에서 불리할 수 있음.

## 2.8 기타 위험
- **고용 및 기술 변화**: GenAI의 도입으로 금융업 내 많은 작업이 자동화되면서 향후 고용에 부정적 영향을 미칠 가능성.
- **환경적 영향**: GenAI 모델의 훈련 및 유지에 드는 막대한 에너지 소비로 인해 환경적 지속 가능성 문제가 제기됨.

# 3장: 정책적 고려 사항

## 3.1 OECD AI 원칙 및 국제적 노력
- 2019년 채택된 **OECD AI 원칙**은 금융에서 신뢰할 수 있는 AI의 책임 있는 관리 기준을 제시하는 국제 표준입니다. 이 원칙은 AI 시스템이 인권, 프라이버시, 공정성, 안전성 등의 잠재적 위험을 다루고 있음을 인식합니다.
- **EU AI Act**: 2021년에 제안된 이 법안은 AI의 안전한 개발을 위한 포괄적인 입법적 프레임워크로, 특히 금융 분야에서 AI 사용을 규제합니다.
- **G7 히로시마 프로세스**: 선진 AI 시스템을 개발하는 조직을 위한 국제적인 원칙을 환영하며, AI 시스템 개발을 감독하는 행동 강령을 수립합니다.

## 3.2 정책적 고려 사항과 잠재적 조치
GenAI 사용이 증가함에 따라 기존의 AI 관련 위험이 악화될 수 있으며, 특히 **딥페이크**와 같은 새로운 위험도 등장합니다.

### 데이터 관리 강화
- 금융에서 GenAI 모델의 신뢰성과 데이터 관리가 중요합니다. **데이터 품질**과 **프라이버시**, **데이터의 적합성**을 보장하기 위한 데이터 거버넌스 관행을 강화해야 합니다.
- **데이터 투명성**을 강화하고, 모델에 사용된 데이터에 대한 문서화 및 보고 체계를 갖추는 것이 중요합니다.

### 모델 거버넌스 및 설명 가능성
- GenAI의 **설명 가능성 부족**은 금융 서비스에서 특히 문제가 될 수 있으며, 이는 대출 거절과 같은 상황에서 고객에게 결과를 설명하기 어렵게 만듭니다. 기존의 AI 모델과 마찬가지로 GenAI 모델도 **투명성**과 **설명 가능성**을 강화할 필요가 있습니다.
- **모델 거버넌스**는 명확한 책임과 감독 체계를 포함해야 하며, 금융 사용 사례에서 발생할 수 있는 위험을 최소화하는 것이 중요합니다.

### 인간 중심적 접근 촉진
- AI 기반 의사결정에서 인간이 개입할 수 있는 여지를 두고, 특히 고위험 금융 서비스에서는 AI 시스템에 과도하게 의존하지 않도록 해야 합니다.
- 고객은 AI의 사용에 대해 명확하게 인지하고, 필요 시 **인간의 개입**을 요청할 수 있어야 합니다.

### 연구 및 역량 강화
- AI 기술의 급격한 발전 속도를 따라잡기 위해 연구 및 개발에 투자하고, 금융 시장 참여자와 정책 입안자들을 교육하여 GenAI 모델의 위험을 관리할 수 있는 도구를 개발해야 합니다.
  
**정책적 결론**:
- **투명성**, **모델 안전성** 강화, **시장 조작** 방지 등의 영역에서 정책적 대응이 필요하며, 이는 미래에도 적용 가능한 정책이 되어야 합니다.
